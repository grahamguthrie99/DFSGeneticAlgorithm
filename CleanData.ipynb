```python
def mergeSources(fantasy_nerd, fantasy_fuel):
    fantasy_fuel['Player Name'] = fantasy_fuel['first_name'] + " " + fantasy_fuel['last_name']
    fantasy_fuel.drop(['first_name', 'last_name', 'ppg_actual', 'value_actual', 'slate'] , axis=1, inplace=True)
    fantasy_nerd.drop(['Likes', 'Opp', 'Team', 'Inj', 'USG', "PER", 'PS'], axis=1, inplace=True)
    merged = pd.merge(left=fantasy_fuel, right=fantasy_nerd, left_on="Player Name", right_on='Player Name')
    merged.set_index('Player Name', inplace=True)
    return merged

def saveMergedFile_toCSV(file):
    game_date = file['game_date'].iloc[0]
    file_name = 'HistoricalData_Merged/{}'.format(game_date)
    file.to_csv(file_name)
    
def mergeAllDates():
    for game_day in dates:
        dfn_date = 'HistoricalData_DFN/DFN NBA DK {}.csv'.format(game_day[0])
        ff_date = 'HistoricalData_FantasyFuel/DFF_NBA_cheatsheet_{}.csv'.format(game_day[1])
        daily_fantasy_nerd_data = pd.read_csv(dfn_date)
        fantasy_fuel_data = pd.read_csv(ff_date)
        merged_file = mergeSources(daily_fantasy_nerd_data, fantasy_fuel_data)
        saveMergedFile_toCSV(merged_file)
```

```python
def buildNBAPlayerSet():
    set_of_players = set()
    for game_day in dates: 
        file_name = 'HistoricalData_Merged/{}'.format(game_day[1])
        df = pd.read_csv(file_name)
        list_of_players = df['Player Name'].tolist()
        for player in list_of_players:
            set_of_players.add(player)
    return set_of_players

def buildPlayerSeasonHistory(player_set):
    for player in list(player_set):
        player_name = player
        df = pd.DataFrame()
        for game_day in dates:
            file_name = 'HistoricalData_Merged/{}'.format(game_day[1])
            new_df = pd.read_csv(file_name)
            new_df.set_index('game_date', inplace=True)
            add_row = new_df.loc[new_df['Player Name'] == player_name]
            df = df.append(add_row, sort=True)
        df = df.sort_index(axis=0 ,ascending=True)
        savePlayerFile_toCSV(df, player_name)
        
def savePlayerFile_toCSV(file, player_name):
    file_name = 'Players/{}'.format(player_name)
    file.to_csv(file_name)
        
        
    
```
